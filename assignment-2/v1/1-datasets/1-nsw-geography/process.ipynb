# NSW Local Health Districts and Air Quality Recording Sites | Processing

The main tasks completed to clean and preprocess this dataset were:

**Data Manipulation**
1. Rename columns.

**Data Cleaning**
1. Remove empty rows.
2. Remove 'WKT' Column.
3. Remove irrelevant row.

Execute the following in the jupyter notebook before execution to ensure that the required libraries are imported:

```python
import pandas as pd
```

## Load Dataset

```python
# File path.
file_path = 'data-raw.csv'

# Read the file.
df = pd.read_csv(file_path)
```

## Data Manipulation

```python
# Rename columns.
df = df.rename(columns={
    'name': 'suburb',
    'description': 'lhd'
})
```

## Data Cleaning

```python
# Remove empty rows.
df = df.dropna()

# Remove 'WKT column.
df = df.drop(columns=['WKT'])

# Remove out-of-scope suburb and Local Health District.
df = df.drop(df[df['suburb'] == 'Albury'].index)
```

## Output Processed Dataset

```python
# File path.
file_path_output = 'data-processed.csv'

# Save the file.
df.to_csv(file_path_output, index=False)
```
